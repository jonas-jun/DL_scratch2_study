```python
import sys
sys.path.append('/Users/jonas/dl_scratch2')
from common.trainer import Trainer
from common.optimizer import Adam
from simple_cbow import SimpleCBOW
from common.util import preprocess, create_contexts_target, convert_one_hot
```

```python
window_size = 1
hidden_size = 5
batch_size = 3
max_epoch = 1000
```

```python
text = 'You say goodbye and I say hello.'
corpus, wtoi, itow = preprocess(text)
```

```python
vocab_size = len(wtoi)
contexts, target = create_contexts_target(corpus, window_size)
target = convert_one_hot(target, vocab_size)
contexts = convert_one_hot(contexts, vocab_size)
```

```python
model = SimpleCBOW(vocab_size, hidden_size)
optimizer = Adam()
trainer = Trainer(model, optimizer)
```

```python
print(target.shape, contexts.shape)
```

```python
trainer.fit(contexts, target, max_epoch=max_epoch, batch_size=batch_size)
trainer.plot()
```

```python
word_vecs = model.word_vecs
for word_id, word in itow.items():
    print(word, word_vecs[word_id])
```
