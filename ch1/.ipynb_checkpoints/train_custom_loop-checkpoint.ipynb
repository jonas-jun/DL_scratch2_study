```python
import sys
sys.path.append('/Users/jonas/DL_scratch2') # use parents directory files
import numpy as np
from optimizer import SGD
from dataset import spiral
import matplotlib.pyplot as plt
from two_layer_net import TwoLayerNet
```

```python
# hyperparameters
max_epoch = 300
batch_size = 30
hidden_size = 10
learning_rate = 1.0
```

```python
# read datasets
x, t = spiral.load_data() # x: input (2 features), t: target (3 classes)
model = TwoLayerNet(input_size=2, hidden_size=hidden_size, output_size=3)
optimizer = SGD(lr=learning_rate)
print(x.shape, t.shape)
```

```python
# training parameters
data_size = len(x)
max_iters = data_size // batch_size
total_loss = 0
loss_count = 0
loss_list = list()

# 300 epochs, 10 iters in an epoch
for epoch in range(max_epoch):
    idx = np.random.permutation(data_size) # permutation: change order randomly
    x = x[idx]
    t = t[idx]
    
    for iters in range(max_iters):
        batch_x = x[iters*batch_size:(iters+1)*batch_size]
        batch_t = t[iters*batch_size:(iters+1)*batch_size]
        
        # update weights
        loss = model.forward(batch_x, batch_t)
        model.backward()
        optimizer.update(model.params, model.grads)
        
        total_loss += loss
        loss_count += 1
        
        # print result every 10 iterations
        if (iters+1) % 10 == 0:
            avg_loss = total_loss / loss_count
            print('Epoch {} | Iters {} / {} | loss {:0.02f}'.format(epoch+1, iters+1, max_iters, avg_loss))
            loss_list.append(avg_loss)
            total_loss, loss_count = 0, 0
        
```
